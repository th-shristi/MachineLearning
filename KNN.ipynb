```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
```

```python
wine = load_wine()
X = wine.data
y = wine.target
```

```python
df = pd.DataFrame(X, columns=wine.feature_names)
df['target'] = y
```

```python
df.head()
```

```python
df.describe()
```

```python
df['target'].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
def predict_wine_class(model, input_features):
    input_df = pd.DataFrame([input_features], columns=wine.feature_names)
    prediction = model.predict(input_df)[0]
    prediction_proba = model.predict_proba(input_df)[0]
    result = wine.target_names[prediction]
    return result, prediction_proba
```

```python
example_input = {
    'alcohol': 13.5,
    'malic_acid': 2.34,
    'ash': 2.36,
    'alcalinity_of_ash': 19.6,
    'magnesium': 106.0,
    'total_phenols': 2.80,
    'flavanoids': 3.24,
    'nonflavanoid_phenols': 0.30,
    'proanthocyanins': 1.83,
    'color_intensity': 5.70,
    'hue': 1.03,
    'od280/od315_of_diluted_wines': 3.15,
    'proline': 1185.0
}
```

```python
result, prediction_proba = predict_wine_class(knn, example_input)
print(f"Prediction: {result}")
print(f"Prediction Probabilities: {prediction_proba}")
```
